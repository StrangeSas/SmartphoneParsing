```python
"""

Для работы с датасетом - названия телефонов были заменены на кампании, что их представляют.

Большинство из телефонов на рынке - представляет собой Андроид различных версий, с разными
характеристики под различные нужды владельца.

Наиболее от своих конкурентов отличается кампания Cubot, что предлагает за низкие цены
наиболее сильное железо в их телефонах

Xiaomi держит за собой рынок среднего сегмента стоимости телефонов, совместно с кампанией Cubot,
однако Cubot имеет за собой большинство продаж ещё и в рынке дешёвого сегмента.

Наиболее дорогостоящие телефоны имеют Apple, как и большое количество покупок, по соотношению
Цена\Качество\Железо телефона.

Ulefone имеет наихудшие оценки, по отношению к своим конкурентам на рынке

"""
```

```python
#Функция для перебора продуктов по ценам и кампаниям
def show_difference_in_price_with_companies(df):
    companies = df['name'].unique()
    for n in companies:
        #Копирование датасета по кампаниям
        df_com = df[df["name"] == n]
        print("Количество Телефонов от данной кампании:")
        print("{}: {}".format(n,len(df_com[df_com["name"] == n]))),
        sns.scatterplot(df_com[(df_com["name"] == n)],y='sales', x="price",alpha=0.9,label=str(n)).set(title='Цены кампаний')
        plt.legend()
    plt.show()#Сделано для отображения всех графиков сразу, а не только одного
```

```python
show_difference_in_price_with_companies(df)
```

```python
"""

Как можно увидеть из графика - Apple имееют наибольшую стоимость и достаточно высокие продажи,однако
Cubot, а так-же Xiaomi в количестве продаж не отстают от Apple

Google - имеет наименьшее количество продаж, по отношению к остальным кампаниям, что представлены 
на графиках

"""
```

```python
#Функция для перебора продуктов по обзорам, ценам и кампаниям
def show_difference_in_sales_with_companies(df):
    companies = df['name'].unique()
    for n in companies:
        #Копирование датасета по кампаниям
        df_com = df[df["name"] == n]
        print("Количество Телефонов с различным количеством отзывов:")
        print("Телефон {}: количество отзывов {}".format(n,len(df_com[df_com["name"] == n]))),
        sns.scatterplot(df_com[(df_com["name"] == n)],y='review_num', x="sales",alpha=0.9,label=str(n)).set(title='Цены кампаний')
        plt.legend()
    plt.show()#Сделано для отображения всех графиков сразу, а не только одного
    
```

```python
show_difference_in_sales_with_companies(df)#Функция для перебора продуктов по ОЗУ, ценам и кампаниям
```

```python
"""

Из данного графика становится видно, что наибольшее влияние на продажи - являет собой именно
качество рекламного образа или то, насколько общественные массы ознакомленны с продуктом.
Это значительно сказывается на продажах

Некоторым отклонением от нормы является Apple, ибо он являет собой коллоса в сфере влияния
на рынке, из-за крайне успешных рекламных кампаний в прошлом, а так-же значительной работы
с общественностью и состроения вокруг собственных продуктов культа.

"""
```
